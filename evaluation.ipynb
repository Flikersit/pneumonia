```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
acc_train = r'C:\Users\User\Desktop\pneumonia\acc_train_pneumania.pkl'
with open(acc_train, 'rb') as file:
    data = pickle.load(file)

x = np.linspace(1, len(data), len(data)) 
plt.plot(x[:200], data[:200])
plt.grid(True)
plt.title("Accuracy train")
plt.show()
```
